# Package

```python
import pandas as pd
from random import sample
import matplotlib.pyplot as plt
```

# Sample

```python
# Simulate a fair dice
# Create sample space
sample_space = [1,2,3,4,5,6]
sample_space
```

```python
# Sample
# sample(sample space, sample size k)
sample(sample_space, 2)
```

# Probability Distribution

```python
probability = [1/6]*6
probability
variable = ['Point 1', 'Point 2', 'Point 3', 'Point 4', 'Point 5', 'Point 6']
```

```python
# Visualization
df = pd.DataFrame({"Point":variable,
                  "Probability":probability})

fig, ax = plt.subplots(figsize=(10,6))


ax.bar('Point','Probability', data=df, color='#009393')
ax.set_xlabel('Outcome');ax.set_ylabel('Probability')
ax.set_title('Probability Distribution of a Fair Dice.')
# Hide the right and top spines
ax.spines[['right', 'top']].set_visible(False)

plt.show()
```
